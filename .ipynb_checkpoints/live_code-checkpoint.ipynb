```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(5, 8, 5,dilation=4,padding="same")
        torch.nn.init.xavier_uniform_(self.conv1.weight)
        self.pool = nn.MaxPool2d(3, 2)
        self.pool2 = nn.AvgPool2d(3, 2)
        
        self.conv2 = nn.Conv2d(8, 16, 5,dilation=3,padding="same")
        torch.nn.init.xavier_uniform_(self.conv2.weight)
        
        self.conv3 = nn.Conv2d(16, 24, 3,dilation=2)
        self.conv4 = nn.Conv2d(24, 48, 2)
        self.conv5 = nn.Conv2d(48, 64, 2)
        self.conv6 = nn.Conv2d(64, 128, 1)
        
        self.fc1 = nn.Linear(1152, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, 16)
        self.fc6 = nn.Linear(16, 8)
        
        self.batch1=nn.BatchNorm2d(8)
        self.batch2=nn.BatchNorm2d(16)
        self.batch3=nn.BatchNorm2d(24)
        self.batch4=nn.BatchNorm2d(64)
        
        self.dropout1=nn.Dropout2d(p=0.2)
        

    def forward(self, x):
        x = self.dropout1(self.batch1(F.relu(self.conv1(x))))
        x = self.pool(self.batch2(F.relu(self.conv2(x))))
        x = self.pool2(self.batch3(F.relu(self.conv3(x))))
        x = F.relu(self.conv4(x))
        x = F.relu(self.conv5(x))
        x = F.relu(self.conv6(x))
        
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = self.fc5(x)
        x = self.fc6(x)
        return x
    
class LiveTrader():
    def __init__(self):
        self.model = Net()
        checkpoint = load_model("experiment://c9d790db2d0544c7a0fcd25ae43ecb90/4_layer_w_volume_classic_CNN_mse_target3")
        self.model.load_state_dict(checkpoint['model_state_dict'])
        epoch = checkpoint['epoch']
        self.model.to(device)

        self.open_position=False
        self.position_history = []
        self.output_history = []
        self.current_position={}
        self.pos_dict_open = {}
        self.pos_dict_closed={}
        self.odf_s = np.array([0.00088979, 0.00134634, 0.00088979, 0.00134634, 0.00088979,
        0.00134634, 0.00088979, 0.00134634])
        self.odf_m = np.array([-0.00102187,  0.00315814, -0.00102187,  0.00315814, -0.00102187,
         0.00315814, -0.00102187,  0.00315814])
        self.ohlc_mean=27993.054512542625
        self.ohlc_std=17832.786089817815
        self.vol_mean=248.0463029065656
        self.vol_std=426.867918968892
        self.long_short=0
        self.verbose=2

    def live_call(self,price):
        print(f"Open Position is {self.open_position}")
        if self.open_position:
            self.in_position()
        else:
            self.out_of_position(price)
        
    def out_of_position(self):
        print("Evaluating Position")
        market = 'BTCUSDT'
        tick_interval = '1m'

        url = 'https://fapi.binance.com/fapi/v1/klines?symbol='+market+'&interval='+tick_interval+'&limit=1025'
        data = requests.get(url).json()
        df = pd.DataFrame(data,columns=cols)
        
        current_ts = df.iloc[-1].copy(deep=True)
        df = df.iloc[:-1]
        print(f"Fetched {len(df)} rows")
        pos_dict_open = {}
        o1 = np.zeros((5,32,32))
        vals = df[["Open","High","Low","Close","Volume"]].values
        for j in range(len(data)):
            o1[:,locs[j][0],locs[j][1]] = vals[j]
        print("Rearranged to hilbert curve")
        ohlc = o1[:4,:,:]
        vol = o1[-1:,:,:]
        ohlc = (ohlc-self.ohlc_mean)/self.ohlc_std
        vol = (vol-self.vol_mean)/self.vol_std
        print("Normalized")
        on1 = torch.Tensor(np.concatenate((ohlc,vol),axis=0).reshape((1,*o1.shape)))
        output=model(on1.to(device)).tolist()[0]
        print(f"Predicted Tensor of shape {on1.shape}")
        
        output_history.append(output)
        short = min(output[::2])
        long = max(output[1::2])
        short_index = output.index(short)//2
        long_index = output.index(long)//2
        norm_max = np.abs((np.array(output) - odf_m)/odf_s).argmax()
        if norm_max%2:
            self.long_short=1
        else:
            self.long_short=-1
        print(f"Long Short Direction {self.long_short}")
        if self.long_short==1:
            print(i)
            open_position=True
            long=long*1.25
            if long_index==0:
                stop_loss=output[0]/5
            else:
                stop_loss = min(output[::2][:long_index])/5
            self.pos_dict_open["status"]="Opened"
            self.pos_dict_open["time"]=current_ts["OpenTime"]
            self.pos_dict_open["price"]=current_ts["Close"]
            self.pos_dict_open["type"]="long"
            self.current_position["type"]="long"
            self.pos_dict_open["take_profit"]=lv*(1+long)
            self.current_position["take_profit"]=lv*(1+long)
            self.pos_dict_open["stop_loss"]=lv*(1+(stop_loss))
            self.current_position["stop_loss"]=lv*(1+(stop_loss))
            self.pos_dict_open["take_profit_ratio"]=(1+long*margin)
            self.current_position["take_profit_ratio"]=(1+long*margin)
            self.pos_dict_open["stop_loss_ratio"]=(1+(stop_loss)*margin)
            self.current_position["stop_loss_ratio"]=(1+(stop_loss)*margin)
            self.pos_dict_open["result"]=None
            if self.verbose==2:
                print("\nOpened Position",self.pos_dict_open)
        
        elif self.long_short==-1:
            print(i)
            open_position=True
            short=short*1.25
            
            if short_index == 0:
                stop_loss = output[1]/6
            else:
                stop_loss = max(output[1::2][:short_index])/6
            
            self.pos_dict_open["status"]="Opened"
            self.pos_dict_open["time"]=current_ts["OpenTime"]
            self.pos_dict_open["price"]=current_ts["Close"]
            self.pos_dict_open["type"]="short"
            self.current_position["type"]="short"
            self.pos_dict_open["take_profit"]=lv*(1+short)
            self.current_position["take_profit"]=lv*(1+short)
            self.pos_dict_open["stop_loss"]=lv*(1+(stop_loss))
            self.current_position["stop_loss"]=lv*(1+(stop_loss))
            self.pos_dict_open["take_profit_ratio"]=(1-short*margin)
            self.current_position["take_profit_ratio"]=(1-short*margin)
            self.pos_dict_open["stop_loss_ratio"]=(1-(stop_loss)*margin)
            self.current_position["stop_loss_ratio"]=(1-(stop_loss)*margin)
            self.pos_dict_open["result"]=None
            if self.verbose==2:
                print("\nOpened Position",self.pos_dict_open)

        self.position_history.append(self.pos_dict_open)
        self.pos_dict_open={}
        
    def in_position(self,price):
        current = price
        if current_position["type"]=="long":
            if current["k"]["l"]<self.current_position["stop_loss"]:
                self.pos_dict_closed["status"]="Closed"
                self.pos_dict_closed["time"]=current["k"]["t"]
                self.pos_dict_closed["price"]=current["k"]["l"]
                self.pos_dict_closed["result"] = "loss"
                profit_margin*=self.current_position["stop_loss_ratio"]*0.99985

                open_position=False
            elif current["k"]["h"]>self.current_position["take_profit"]:
                self.pos_dict_closed["status"]="Closed"
                self.pos_dict_closed["time"]=current["k"]["t"]
                self.pos_dict_closed["price"]=current["k"]["h"]
                self.pos_dict_closed["result"] = "profit"
                profit_margin*=self.current_position["take_profit_ratio"]*0.99985
                open_position=False
        elif current_position["type"]=="short":
            if current["k"]["h"]>self.current_position["stop_loss"]:
                self.pos_dict_closed["status"]="Closed"
                self.pos_dict_closed["time"]=current["k"]["t"]
                self.pos_dict_closed["price"]=current["k"]["h"]
                self.pos_dict_closed["result"] = "loss"
                profit_margin*=self.current_position["stop_loss_ratio"]*0.99985
                open_position=False
            elif current["k"]["l"]<self.current_position["take_profit"]:
                self.pos_dict_closed["status"]="Closed"
                self.pos_dict_closed["time"]=current["k"]["t"]
                self.pos_dict_closed["price"]=current["k"]["l"]
                self.pos_dict_closed["result"] = "profit"
                profit_margin*=self.current_position["take_profit_ratio"]*0.99985

                open_position=False

        
        if not self.open_position:
            self.pos_dict_closed["type"] = self.current_position["type"]
            self.pos_dict_closed["take_profit"] = self.current_position["take_profit"]
            self.pos_dict_closed["stop_loss"] = self.current_position["stop_loss"]
            self.pos_dict_closed["take_profit_ratio"] = self.current_position["take_profit_ratio"]
            self.pos_dict_closed["stop_loss_ratio"] = self.current_position["stop_loss_ratio"]
            self.position_history.append(self.pos_dict_closed)
            if self.verbose==2:
                print("\nClosed Position", self.pos_dict_closed)
            self.pos_dict_closed={}
            self.current_position={}
    
```
